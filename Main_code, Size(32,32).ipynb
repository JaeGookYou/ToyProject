```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from datetime import datetime
from tensorflow.keras.preprocessing.text import text_to_word_sequence
from sklearn.model_selection import train_test_split
import random
```

```python
#파일 생성
os.mkdir('data/new_data')

#파일 이름 리스트화
dw_img_list=os.listdir('data/dwenjang/')
kim_img_list=os.listdir('data/kimchi/')
soon_img_list=os.listdir('data/soondubu/')
```

```python
#32*32로 파일 크기 변환(된장찌개)
count = 0
err_count = 0
for idx in range(1,len(dw_img_list)):
    dw_fname=dw_img_list[idx]
    dw_img_path=os.path.join('data/dwenjang/', dw_fname)
    
    dw_img=cv2.imread(str(dw_img_path))
    try:
        dw_res=cv2.resize(dw_img, dsize=(32, 32), interpolation=cv2.INTER_CUBIC)
        cv2.imwrite('data/new_data/dwenjang_{}.jpg'.format(idx),dw_res)
        print("사진 변환 :", idx)
        count += 1
    except:
        err_count += 1
        print("사진 변환 오류 : ", err_count)
print('--------------------------------------')
print("사진 변환 성공 횟수 : ", count)
print("사진 변환 오류 횟수 : ", err_count)
```

```python
#32*32로 파일 크기 변환(김치찌개)
count = 0
err_count = 0
for idx in range(1,len(kim_img_list)):
    kim_fname=kim_img_list[idx]
    kim_img_path=os.path.join('data/kimchi/', kim_fname)
    
    kim_img=cv2.imread(str(kim_img_path))
    try:
        kim_res=cv2.resize(kim_img, dsize=(32, 32), interpolation=cv2.INTER_CUBIC)
        cv2.imwrite('data/new_data/kimchi_{}.jpg'.format(idx),kim_res)
        print("사진 변환 :", idx)
        count += 1
    except:
        err_count += 1
        print("사진 변환 오류 : ", err_count)
print('--------------------------------------')
print("사진 변환 성공 횟수 : ", count)
print("사진 변환 오류 횟수 : ", err_count)
```

```python
#32*32로 파일 크기 변환(순두부찌개)
count = 0
err_count = 0
for idx in range(1,len(soon_img_list)):
    soon_fname=soon_img_list[idx]
    soon_img_path=os.path.join('data/soondubu/', soon_fname)
    
    soon_img=cv2.imread(str(soon_img_path))
    try:
        soon_res=cv2.resize(soon_img, dsize=(32, 32), interpolation=cv2.INTER_CUBIC)
        cv2.imwrite('data/new_data/soondubu_{}.jpg'.format(idx),soon_res)
        print("사진 변환 :", idx)
        count += 1
    except:
        err_count += 1
        print("사진 변환 오류 : ", err_count)
print('--------------------------------------')
print("사진 변환 성공 횟수 : ", count)
print("사진 변환 오류 횟수 : ", err_count)
```

```python
#데이터 순서 섞기
data_list=os.listdir('data/new_data/')
random.shuffle(data_list)
print(data_list)
```

```python
#데이터 라벨화
lab=[]

for i in range(len(data_list)):
    flabel=text_to_word_sequence(data_list[i])
    if flabel[0]=='dwenjang':
        lab.append(0)
    
    elif flabel[0]=='kimchi':
        lab.append(1)
        
    elif flabel[0]=='soondubu':
        lab.append(2)
        
label=np.array(lab)
print(label)
```

```python
#이미지 데이터 합치기
images=np.zeros((len(data_list),32,32,3))

for i, n in enumerate(data_list[:400]):
    path='data/new_data'
    real_path=os.path.join(path,data_list[i])
    image=cv2.imread(real_path)/255.
    images[i,:,:,:]=image

print(images.shape,label.shape)
```

```python
#train, test 데이터 분류
x=images
y=label
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=359)
print(x_train.shape,y_train.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#학습 모델 구축
cnn = Sequential()

cnn.add(Conv2D(32,(3,3), activation = 'relu', padding = 'same', input_shape = (32, 32, 3)))
cnn.add(Conv2D(32,(3,3), activation = 'relu', padding = 'same'))
cnn.add(MaxPooling2D(pool_size = (2,2)))
cnn.add(Dropout(0.25))

cnn.add(Conv2D(64,(3,3), activation = 'relu', padding = 'same'))
cnn.add(Conv2D(64,(3,3), activation = 'relu', padding = 'same'))
cnn.add(MaxPooling2D(pool_size = (2,2)))
cnn.add(Dropout(0.25))

cnn.add(Conv2D(128,(3,3), activation = 'relu', padding = 'same'))
cnn.add(MaxPooling2D(pool_size = (2,2)))
cnn.add(Dropout(0.25))

cnn.add(Conv2D(128,(3,3), activation = 'relu', padding = 'same'))
cnn.add(MaxPooling2D(pool_size = (2,2)))
cnn.add(Dropout(0.25))

cnn.add(Conv2D(256,(3,3), activation = 'relu', padding = 'same'))
cnn.add(MaxPooling2D(pool_size = (2,2)))
cnn.add(Dropout(0.25))

cnn.add(Flatten())

cnn.add(Dense(128, activation = 'relu'))
cnn.add(Dropout(0.5))
cnn.add(Dense(10, activation = 'softmax'))

cnn.summary()
```

```python
#데이터 학습
from datetime import datetime

start_time = datetime.now()

cnn.compile(loss = 'sparse_categorical_crossentropy', optimizer = Adam(), metrics = ['accuracy'])

hist = cnn.fit(x_train, y_train, batch_size = 256, epochs = 40, validation_data = (x_test, y_test))

end_time = datetime.now()

print('Elapsed Time => ', end_time - start_time)
```

```python
cnn.evaluate(x_test, y_test)
```

```python
#정확도
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Accuracy Trend')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc = 'best')
plt.grid()
plt.show()
```

```python
#손실값
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Loss Trend')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc = 'best')
plt.grid()
plt.show()
```
